```python
def area(p,q,r):
  return (r[1]-p[1])*(q[0]-p[0])-(q[1]-p[1])*(r[0]-p[0])


def intersection(p,q,r,s):
  if area(p,q,r)*area(p,q,s)<0 and area(r,s,p)*area(r,s,q)<0:
    return True
  elif area(p,q,r)*area(p,q,s)==0 or area(r,s,p)*area(r,s,q)==0:
    if (r[0]<=max(p[0],q[0]) and r[0]>=min(p[0],q[0])) or (s[0]<=max(p[0],q[0]) and s[0]>=min(p[0],q[0])) or (p[0]<=max(r[0],s[0]) and p[0]>=min(r[0],s[0])) or (q[0]<=max(r[0],s[0]) and q[0]>=min(r[0],s[0])):
      return True
    else:
      return False

  else:
    return False

def convex(p,q,r):
  if area(p,q,r)>0:
    return True
  else:
    return False



```

```python
import math
import matplotlib.pyplot as plt


input_sizes = [10, 100, 1000]


log_input_sizes = [math.log(n) for n in input_sizes]


t_n2 = [0.25, 1.1086, 19.2919]  # Times for O(n^2)
t_n3 = [0.25, 21, 3253]         # Times for O(n^3)


plt.figure(figsize=(8, 6))
plt.plot(log_input_sizes, t_n2, 'r-', label="O(n^2)", linewidth=2)
plt.plot(log_input_sizes, t_n3, 'b-', label="O(n^3)", linewidth=2)


plt.xlabel('log(Input Size)', fontsize=12)
plt.ylabel('Execution Time (seconds)', fontsize=12)
plt.title('Execution Time vs log(Input Size)', fontsize=14)


plt.legend()


plt.grid(True)
plt.show()
```
